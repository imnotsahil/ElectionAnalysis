```python
##---------Type the code below this line------------------##
from sklearn.feature_selection import mutual_info_classif, chi2
from sklearn.preprocessing import LabelEncoder

# Preparing the data
X = data.drop('Party', axis=1).select_dtypes(include=['float64', 'int64'])  # features
y = LabelEncoder().fit_transform(data['Party'])  # encoded target

# Calculating Mutual Information
mi_scores = mutual_info_classif(X, y)
mi_scores_series = pd.Series(mi_scores, index=X.columns).sort_values(ascending=False)

# Calculating Chi-Squared scores (only non-negative values)
non_negative_cols = X.columns[(X >= 0).all()].tolist()
chi2_scores, chi2_p_values = chi2(X[non_negative_cols], y)
chi2_scores_series = pd.Series(chi2_scores, index=non_negative_cols).sort_values(ascending=False)

print("Top 5 features based on Mutual Information:")
print(mi_scores_series[:5])
print("\nTop 5 features based on Chi-Squared Test:")
print(chi2_scores_series[:5])


```

## 6.1 ML technique 1 + Justification


1) Classification using Decision Tree Classifier: Given that we have the 'Party' column as a categorical target, a classification algorithm is suitable. We'll predict the party based on the other features.


--Interpretability: One of the significant advantages of Decision Trees is their interpretability. You can visualize the tree and understand the decisions it's making, which is beneficial for political datasets where understanding the reasoning can be crucial.

--Flexibility: Decision Trees can handle both numerical and categorical data. Our dataset comprises both types of data.

--Feature Importance: Decision Trees naturally provide feature importance, which can help in understanding which factors are most influential in determining the party.





```python
##---------Type the code below this line------------------##

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report



# Selecting features and target variable
X_classification = data[['Votes', 'Percentage', 'Hondt']]
y_classification = data['Party']

# Splitting data
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_classification, y_classification, test_size=0.2, random_state=42)

# Initializing and training
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train_class, y_train_class)

# Making predictions
predictions_class = dt_classifier.predict(X_test_class)

# Calculating metrics
accuracy_class = accuracy_score(y_test_class, predictions_class)
precision_class = precision_score(y_test_class, predictions_class, average='weighted', zero_division=1)
recall_class = recall_score(y_test_class, predictions_class, average='weighted', zero_division=1)
f1_class = f1_score(y_test_class, predictions_class, average='weighted')

# Printing metrics
print(f"Accuracy: {accuracy_class:.4f}")
print(f"Precision (Weighted): {precision_class:.4f}")
print(f"Recall (Weighted): {recall_class:.4f}")
print(f"F1 Score (Weighted): {f1_class:.4f}")
print("\nClassification Report for each Party:")
print(classification_report(y_test_class, predictions_class))

```
